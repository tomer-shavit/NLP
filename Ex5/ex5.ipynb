```python
# Q2: Dependency-based Extractor
import wikipedia
import spacy

nlp = spacy.load("en_core_web_sm")

page_title_q2 = "Brad Pitt"
page_content_q2 = wikipedia.page(page_title_q2, auto_suggest=False).content
doc_q2 = nlp(page_content_q2)

proper_noun_heads = []
for token in doc_q2:
    if token.pos_ == "PROPN" and token.dep_ != "compound":
        proper_noun_heads.append(token)

proper_noun_map = {}
for head in proper_noun_heads:
    pn_list = [head.text]
    for child in head.children:
        if child.dep_ == "compound" and child.pos_ == "PROPN":
            pn_list.insert(0, child.text)
    proper_noun_map[head] = " ".join(pn_list)

def condition_1(h1, h2):
    return (
        h1.head == h2.head
        and h1.dep_ == "nsubj"
        and h2.dep_ == "dobj"
    )

def condition_2(h1, h2):
    h = h1.head
    p = h2.head
    return (
        h1.dep_ == "nsubj"
        and h2.dep_ == "pobj"
        and p.dep_ == "prep"
        and p.head == h
    )

triplets_q2 = []
for i in range(len(proper_noun_heads)):
    for j in range(len(proper_noun_heads)):
        if i == j:
            continue
        h1 = proper_noun_heads[i]
        h2 = proper_noun_heads[j]

        if condition_1(h1, h2):
            verb = h1.head.text
            subj_text = proper_noun_map[h1]
            obj_text  = proper_noun_map[h2]
            triplets_q2.append((subj_text, verb, obj_text))
        elif condition_2(h1, h2):
            main_verb = h1.head.text
            prep = h2.head.text
            relation_text = main_verb + " " + prep
            subj_text = proper_noun_map[h1]
            obj_text  = proper_noun_map[h2]
            triplets_q2.append((subj_text, relation_text, obj_text))

print("Q2 Triplets (Dependency-based)")
for subj, rel, obj_ in triplets_q2:
    print(f"Subject: {subj}\tRelation: {rel}\tObject: {obj_}")
```

```python
import wikipedia
import spacy
import random

nlp = spacy.load("en_core_web_sm")

def extract_triplets_pos(doc):
    spans = []
    i = 0
    while i < len(doc):
        if doc[i].pos_ == "PROPN":
            start = i
            while i < len(doc) and doc[i].pos_ == "PROPN":
                i += 1
            end = i
            spans.append((doc[start:end].text, start, end))
        else:
            i += 1

    results = []
    for idx in range(len(spans) - 1):
        subj_span, subj_start, subj_end = spans[idx]
        obj_span, obj_start, obj_end = spans[idx + 1]
        in_between = doc[subj_end:obj_start]

        if any(t.pos_ == "PUNCT" for t in in_between):
            continue
        if not any(t.pos_ == "VERB" for t in in_between):
            continue
        rel_tokens = [t.text for t in in_between if t.pos_ in ["VERB", "ADP"]]
        rel_text = " ".join(rel_tokens).strip()
        if rel_text:
            results.append((subj_span, rel_text, obj_span))
    return results

def extract_triplets_dep(doc):
    heads = [t for t in doc if t.pos_ == "PROPN" and t.dep_ != "compound"]

    pn_map = {}
    for h in heads:
        pn_list = [h.text]
        for child in h.children:
            if child.dep_ == "compound" and child.pos_ == "PROPN":
                pn_list.insert(0, child.text)
        pn_map[h] = " ".join(pn_list)

    def cond_1(h1, h2):
        return (
            h1.head == h2.head
            and h1.dep_ == "nsubj"
            and h2.dep_ == "dobj"
        )

    def cond_2(h1, h2):
        return (
            h1.dep_ == "nsubj"
            and h2.dep_ == "pobj"
            and h2.head.dep_ == "prep"
            and h2.head.head == h1.head
        )

    results = []
    for i in range(len(heads)):
        for j in range(len(heads)):
            if i == j:
                continue
            h1 = heads[i]
            h2 = heads[j]
            if cond_1(h1, h2):
                verb = h1.head.text
                s_text = pn_map[h1]
                o_text = pn_map[h2]
                results.append((s_text, verb, o_text))
            elif cond_2(h1, h2):
                main_verb = h1.head.text
                prep = h2.head.text
                r_text = main_verb + " " + prep
                s_text = pn_map[h1]
                o_text = pn_map[h2]
                results.append((s_text, r_text, o_text))
    return results

evaluation_pages = [
    "Donald Trump",
    "Ruth Bader Ginsburg",
    "J. K. Rowling"
]

print("Q3: Part A - Report total triplets from both extractors")
for page_title in evaluation_pages:
    content = wikipedia.page(page_title, auto_suggest=False).content
    doc_eval = nlp(content)

    q1_trips = extract_triplets_pos(doc_eval)
    q2_trips = extract_triplets_dep(doc_eval)

    print(f"\nPage: {page_title}")
    print(f"  Q1 (POS-based) triplets: {len(q1_trips)}")
    print(f"  Q2 (Dependency-based) triplets: {len(q2_trips)}")

print("Q3: Part B - Randomly Sample 5 Triplets per Page (Manual Check)")

def sample_triplets_by_page(pages, extract_fn, extractor_label):
    print(f"\n--- {extractor_label} Random Samples ---")
    for ptitle in pages:
        docp = nlp(wikipedia.page(ptitle, auto_suggest=False).content)
        all_trips = extract_fn(docp)
        sample_size = min(5, len(all_trips))
        sampled = random.sample(all_trips, sample_size) if sample_size > 0 else []

        print(f"\nPage: {ptitle} - Showing {sample_size} random triplets:")
        for s, r, o in sampled:
            print(f"  (\"{s}\", \"{r}\", \"{o}\")")

sample_triplets_by_page(evaluation_pages, extract_triplets_pos, "Q1 (POS-based)")

sample_triplets_by_page(evaluation_pages, extract_triplets_dep, "Q2 (Dependency-based)")
```
